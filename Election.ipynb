```python
import requests
from bs4 import BeautifulSoup
import pandas as pd
```

```python
superlist = []

for x in range (1,20):

    url = f'https://results.eci.gov.in/ResultAcGenDec2022/statewiseS06{x}.htm'


    page = requests.get(url)


    soup = BeautifulSoup(page.text, 'lxml')
    soup

    table1 = soup.find('table', border = 1)
    table1

    headers = []

    for i in table1.find_all('th'):
        title = i.text
        headers.append(title)

    row2 = []
    for j in table1.find_all('tr')[13:]:
        row_data = j.find_all('td')
        row = [i.text for i in row_data]
        row2.append(row)

    main_list = []
    
    num = 0
    
    while num <= 72:
        sublist = []
        sublist.append(row2[num][0])
        sublist.append(row2[num][1])
        sublist.append(row2[num][2])
        sublist.append(row2[num][4])
        sublist.append(row2[num][15])
        sublist.append(row2[num][17])
        sublist.append(row2[num][28])
        sublist.append(row2[num][29])
        main_list.append(sublist)

        num = num + 9
    
    superlist.extend(main_list)

```

```python
our_data= pd.DataFrame(superlist, columns=headers)
our_data['Margin'] = pd.to_numeric(our_data['Margin'])
```
