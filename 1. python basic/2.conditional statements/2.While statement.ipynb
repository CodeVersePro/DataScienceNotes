### Using if condition to break WHILE loop

```python
initial_speed = 0
top_speed = 182000
safty_speed = 20

while initial_speed <= top_speed:

    print(f'SPEED : {initial_speed}')
    initial_speed = initial_speed + 1

    if initial_speed == safty_speed:
        break
```

### using else statement with while

```python
def road_check(r):
    
    if (r != 1) and (r!=2) and (r!=3):
        
        raise Exception('Please choose valid road type : 1,2 or 3')
        
        
    else:
        
        return r
```

```python
initial_speed = 0
top_speed = 300

## describing type of the road



try:
    
    road_type_key = int(input('Please enter 1 = Highway, 2 = two way, 3 = local one way'))
    road_type = road_check(road_type_key)
    
except ValueError as e:
    
    print('Please enter valid integer input')
    
    
except Exception as e:
    
    print(e)
    

## setting safe speed for different road

safe_speed = 0

if road_type == 1:
    
    safe_speed = 120
    
elif road_type == 2:
    
    safe_speed = 80
    
else:
    
    safe_speed = 60
    
    
## while loop for speed

while initial_speed < top_speed:
    
    print(f'Speed : {initial_speed}')
    initial_speed += 1
    
    if initial_speed > safe_speed:
        print(f'You achieved top speed : {safe_speed}, you can not cross it')
        break
        
    
```
